```python
# 모듈 불러오기
import pandas as pd
from collections import Counter
```

```python
# 데이터 불러오기
drama = pd.read_excel('./justwatch.xlsx', index_col=0)
drama.head()
```

```python
# 데이터 살펴보기
drama.info()
```

```python
drama[drama['season_episode'] >= 2]
```

```python
# 제조 국가
drama['Production country'].value_counts().head(20)
```

```python
# 연도
drama['year'].value_counts().head(20)
```

```python
# 장르
genre = []
for idx, row in drama.iterrows():
    tmp_list= row['genre'].split(', ')
    for x in tmp_list:
        genre.append(x)
```

```python
counts = Counter(genre)
for element, count in counts.items():
    print(f'{element}: {count}')
```
