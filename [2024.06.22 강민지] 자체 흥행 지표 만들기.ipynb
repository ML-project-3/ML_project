## 2. 함수 만들기

### (1) IMDb 평점, 평점 매긴 사람 수

```python
# IMDb 평점
def get_imdb_rate(x):
    imdb_rate_lst = []

    # 딕셔너리에서 평점 부분만 가져옴
    for idx, val in x.items():
        imdb_rate_lst.append(val[0])

    # 시즌 평점 평균 구하기
    imdb_rate = np.mean(imdb_rate_lst)

    return imdb_rate
```

```python
# IMDb 평점 매긴 사람 수
def get_imdb_rate_cnt(x):
    imdb_rate_cnt_lst = []

    # 딕셔너리에서 평점 매긴 사람만 가져옴
    for idx, val in x.items():
        imdb_rate_cnt_lst.append(val[1])


    # 평점을 매긴 사람의 시즌 평균 구하기
    imdb_rate_cnt = np.mean(imdb_rate_cnt_lst)

    return imdb_rate_cnt
```

### (3) 유지도

```python
# 평점 유지도
def get_rate_retention(x):
    imdb_tmp = list(x.values())
    k = len(imdb_tmp)
    score = (imdb_tmp[k-1][0] / imdb_tmp[0][0])
    return score
```

```python
# 평점 매긴 사람 수의 유지도
def get_rate_cnt_retention(x):
    imdb_tmp = list(x.values())
    k = len(imdb_tmp)
    score = (imdb_tmp[k-1][1] / imdb_tmp[0][1])
    return score
```

### (4) 제작 국가

```python
# 제작 국가 순위 별 점수 부여
def get_country_score(x):
    country_tmp = x.split(', ')

    # 국가 순위 별 점수 (TOP 10 사이트 참고)
    dict_score = {'미국': 1, '대한민국': 0.95, '영국': 0.9, '스페인': 0.85, '프랑스': 0.8, '콜롬비아': 0.75, '멕시코': 0.7,
                 '일본': 0.65, '인도': 0.6, '캐나다': 0.55, '독일': 0.5, '폴란드': 0.45, '이탈리아': 0.4, '중국': 0.35, '러시아': 0.3,
                 '브라질': 0.25, '터키': 0.2, '노르웨이': 0.15, '호주': 0.1}

    # 가장 첫 번째 위치한 국가를 기준으로 점수 부여
    score = dict_score.get(country_tmp[0], 0.05)

    return score
```

### (5) 대륙

```python
# 대륙별 국가 리스트 딕셔너리 (플릭스 패트롤 공식 홈페이지 참고)
continent_countries = {
    '북미': ['미국', '캐나다'],
    '유럽': [
        '영국', '스페인', '독일', '프랑스', '폴란드', '이탈리아', '스웨덴', '벨기에', '덴마크', '노르웨이', '룩셈부르크',
        '네덜란드', '러시아', '포르투갈', '아이슬란드', '스위스', '체코 공화국', '핀란드', '그리스', '오스트리아', '아일랜드'
    ],
    '아시아': [
        '대한민국', '일본', '인도', '중국', '대만', '태국', '필리핀', '쿠웨이트', '말레이시아', '인도네시아', 
        '터키', '이스라엘', '요르단', '레바논', '사우디아라비아'
    ],
    '아프리카': ['남아프리카공화국', '나이지리아', '이집트', '케냐'],
    '라틴 아메리카': ['멕시코', '브라질', '콜롬비아', '아르헨티나', '칠레'],
    '오세아니아': ['호주', '뉴질랜드']
}

# 대륙별 점수 딕셔너리 (플랙스 패트롤 공식 홈페이지 참고)
continent_scores = {
    '북미': 0.462,
    '유럽': 0.28,
    '아시아': 0.204,
    '아프리카': 0.016,
    '라틴 아메리카': 0.014,
    '오세아니아': 0.009
}

# 국가 문자열을 받아서 대륙 점수를 반환하는 함수
def get_continent_score(country_str):

    # 국가를 ', '로 분리
    countries = country_str.split(', ')
    
    # 첫 번째 국가를 선택
    first_country = countries[0]
    
    # 국가가 속한 대륙 찾기
    for continent, country_list in continent_countries.items():
        if first_country in country_list:

            # 대륙의 점수 반환
            return continent_scores.get(continent, 0.05)
    
    # 국가가 대륙 리스트에 없는 경우 기본 점수 0.05 반환
    return 0.05
```

### (6) 장르

```python
# 넷플릭스 장르 선호도 순위 리스트 (선호도 순서대로 정렬)
genres = [
    '드라마', '코미디', 'SF', '액션', '스릴러', '범죄', '공포', '로맨스', '판타지', '전쟁', '서부', '그 외'
]

# 각 장르에 점수를 선형적으로 할당하여 딕셔너리로 만들기
genre_scores = {genre: 1 - i / (len(genres)) for i, genre in enumerate(genres)}

# '그 외'에 해당되는 장르 리스트
other_genres = ['역사', '가족', '음악', 'Made in Europe', '스포츠']

# 장르 문자열을 받아서 점수를 계산하는 함수

def get_genre_score(genre_str):
    # 장르를 ', '로 분리
    genre_list = genre_str.split(', ')
    
    # '그 외'에 포함된 하위 장르가 있는지 확인
    for genre in genre_list:
        if genre in other_genres:
            return genre_scores['그 외']
    
    # 각 장르의 점수의 평균 계산
    total_score = 0
    count = 0
    for genre in genre_list:
        score = genre_scores.get(genre, genre_scores['그 외'])  # 없는 장르는 '그 외' 점수로 처리
        total_score += score
        count += 1
    
    # 평균 점수 계산
    average_score = total_score / count if count > 0 else 0
    return average_score
```

```python
# 장르별 점수
imdb['genre'].apply(get_genre_score)
```

```python
# 점수 부여: 인원 수에만 로그를 취함
for i in range(0, 905):
    imdb_tmp = imdb.loc[i, 'season_1']; watcha_tmp = watcha.loc[i, 'season_1']
    genre = imdb.loc[i, 'genre']; country = imdb.loc[i, 'production_country']; age = imdb.loc[i, 'age_rating']

    rate = 0.8 * (get_imdb_rate(imdb_tmp)) + 0.2 * (get_watcha_rate(watcha_tmp))
    cnt = 0.8 * np.log(get_imdb_rate_cnt(imdb_tmp)) + 0.2 * np.log(get_watcha_rate_cnt(watcha_tmp))
    ret = 1/4 * (get_rate_retention(imdb_tmp)) + 3/4 * (get_rate_cnt_retention(imdb_tmp))
    info = 0.4 * (get_country_score(country) + get_continent_score(country)) / 2 + 0.4 * get_genre_score(genre) + 0.1 * get_genre_diversity(genre) + 0.1 * (get_age_rating_score(age))
    total = (0.7 * (rate + 3 * cnt + 1 * ret) + 0.3 * info) * 100

    imdb.loc[i, 'score'] = total
```
